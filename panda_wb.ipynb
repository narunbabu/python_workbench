```python
def adder(ele1,ele2):
   return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
```

```python
df
```

```python
mdf=df.pipe(adder,2)
mdf
```

```python
mdf=df.apply(lambda x: x.max() - x.min())
mdf
rdf=df.applymap(lambda x:x*100)
rdf
```
